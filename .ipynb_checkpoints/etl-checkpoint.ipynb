```python
# imports
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
# Pandas CSV
csv_file = "data/SO_pandas.csv"
so_pandas_df = pd.read_csv(csv_file)
so_pandas_df.head()
```

```python
# Python Answers CSV
csv_file ='data/Answers_small1.csv'
so_python_answers_df = pd.read_csv(csv_file, encoding='utf-8')
so_python_answers_df
```

```python
# Python Questions CSV
csv_file ='data/Questions_small.csv'
so_python_questions_df = pd.read_csv(csv_file, encoding='latin-1')
so_python_questions_df
```

```python
# Join dfs
a = so_python_questions_df.set_index("Id").join(so_python_answers_df.set_index("ParentId"), lsuffix="_question", rsuffix="_answer")
b = a.dropna()
b
```

```python
# Rename columns
dicti = {
    "OwnerUserId_question": "questioner_id",
    "CreationDate_question": "question_asked_date",
    "Score_question": "question_score",
    "Title": "question_title",
    "Body_question": "question_text",
    "Id": "id",
    "OwnerUserId_answer": "answerer_id",
    "CreationDate_answer": "question_answered_date",
    "Score_answer": "answer_score",
    "Body_answer": "answer_text"
}

c = b.rename(columns=dicti)

c
```

```python
# Turn all floats to int
c["questioner_id"] = round(c["questioner_id"].astype(int),0)
c["id"] = round(c["id"].astype(int),0)
c["answerer_id"] = round(c["answerer_id"].astype(int),0)
c["answer_score"] = round(c["answer_score"].astype(int),0)

c.dtypes
```

```python
# Perform L
rds_connection_string = "postgres:postgres@localhost:5432/sodb"
engine = create_engine(f'postgresql://{rds_connection_string}')



c.to_sql(name='so_qa', con=engine, if_exists='append', index=False)
```
